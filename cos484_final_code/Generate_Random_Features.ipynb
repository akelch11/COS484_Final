## Create random explanations

Goal:
- create a script that will output a text file with $n>0$ random explanations
- have '{e1}' and '{e2}' appear in each explanation

Approaches for randomness:
- sample random words from english lanuage 
- sample random words from wikipedia page about marriage (input a text document and sample words from that)
- sample sentences from wikipedia page about marriage (maybe can create these ourselves)


```python
# Get list of 10,000 words
word_site = "https://www.mit.edu/~ecprice/wordlist.10000"

response = requests.get(word_site)
WORDS = response.content.splitlines()

WORDS = [word.decode() for word in WORDS]
```

```python
import random
# denote the number of explanations
N = 40

# make length of explanations vary between 5 and 15 words
L = 3
U = 8

def gen_example(words_list, L, U):
  num_choices = np.random.randint(L, U+1)
  word_choices = random.choices(words_list, k=num_choices)

  pos_1 = np.random.randint(0, len(word_choices)+1)
  pos_2 = np.random.randint(0, len(word_choices)+1)

  word_choices.insert(pos_1, "{e1}")
  word_choices.insert(pos_2, "{e2}")

  ex = ""

  for word in word_choices:
    ex += word + " "
    
  return ex
```

```python
output_dir = './spouse/random_explanations_v1/'
name = "explanations"
ext = '.txt'

path = output_dir + name + ext

output_str = ''

# make output string
for i in range(N):
  if i == N-1:
    output_str += gen_example(WORDS, L, U)
  else:
    output_str += gen_example(WORDS, L, U) + "\n"

with open(path, "w") as text_file:
    text_file.write(output_str)
```
